## ChatBot

```python
# Step 1: simple chat example with a system prompt

import os
from openai import OpenAI


client = OpenAI(
    api_key="sk-dPf9jTD8mhwyk0LKsDxf4xoTSMlStmfNh1ADn63OVWM1G3MF",
    base_url="https://api.gapgpt.app/v1"
)

#api_key = os.getenv("OPENAI_API_KEY")
#client = OpenAI(api_key=api_key)

MODEL = "gapgpt-qwen-3.5"  # Or "gpt-3.5-turbo", "gpt-4o-mini", etc.
TEMPERATURE = 0.7
MAX_TOKENS = 100
SYSTEM_PROMPT = "You are a fed up and sassy assistant who hates answering questions."

def chat(user_input):
    response = client.chat.completions.create(
        model=MODEL,
        messages=[
            {"role": "system", "content": SYSTEM_PROMPT},
            {"role": "user", "content": user_input}
        ],
        temperature=TEMPERATURE,
        max_tokens=MAX_TOKENS
    )

    reply = response.choices[0].message.content
    return reply

# Example usage
print(chat("What is the capital of France?"))
```

```python
# Step 2: simple chat example with a system prompt and conversation history

MODEL = "gapgpt-qwen-3.5"  # Or "gpt-3.5-turbo", "gpt-4o-mini", etc.
TEMPERATURE = 0.7
MAX_TOKENS = 100
SYSTEM_PROMPT = "You are a fed up and sassy assistant who hates answering questions."
messages = [{"role": "system", "content": SYSTEM_PROMPT}]

def chat(user_input):
    messages.append({"role": "user", "content": user_input})

    response = client.chat.completions.create(
        model=MODEL,
        messages=messages,
        temperature=TEMPERATURE,
        max_tokens=MAX_TOKENS
    )

    reply = response.choices[0].message.content
    messages.append({"role": "assistant", "content": reply})

    return reply

while True:
    user_input = input("You: ")
    if user_input.strip().lower() in {"exit", "quit"}:
        break                     # Fixed indentation
    answer = chat(user_input)
    print("Assistant:", answer)   # Now printing the response
```
